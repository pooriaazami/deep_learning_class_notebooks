```python
import tensorflow as tf
import tensorflow.keras as keras
import tensorflow.keras.layers as layers
```

```python
input_layer = layers.Input(shape=(32, 32, 3), name='input_layer')

x = layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', name='conv_1_1')(input_layer)
x = layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', name='conv_1_2')(x)
x = layers.MaxPooling2D(2, name='max_pool_1')(x)

residual_input = layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', name='conv_2_1')(x)
x = layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', name='conv_2_2')(residual_input)
x = layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', name='conv_2_3')(x)

x = layers.Add(name='add_layer')([x, residual_input])
x = layers.MaxPooling2D(2, name='max_pool_2')(x)

x = layers.Flatten(name='flatten')(x)

output_layer = layers.Dense(10, activation='softmax', name='classification_output')(x)
second_output = layers.Dense(20, name='regression_output')(x)

model = keras.Model(inputs=[input_layer], outputs=[output_layer, second_output])
```

```python
keras.utils.plot_model(model, show_shapes=True)
```
